# merceka_core

> Core utilities for merceka projects


## Install

```sh
pip install merceka_core
```


## How to use

### Local Ollama Models

```python
from merceka_core.llm import LLM

llm = LLM("gemma3:27b")
response = llm.generate("Hello, how are you?")
```

### Cloud Models via OpenRouter

```python
from merceka_core.llm import LLM

llm = LLM("openrouter/google/gemini-2.5-flash-lite-preview-09-2025")
response = llm.generate("Hello, how are you?")
```

### Structured Output

```python
from merceka_core.llm import LLM, OutputSchema

class MyResponse(OutputSchema):
    content: str
    sentiment: str

llm = LLM("gemma3:27b", output_schema=MyResponse)
response = llm.generate("I love this!")
print(response.sentiment)  # "positive"
```
